从文本中提取网站地址

```python
import requests
import json
import re,os
#	http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+
def extract_urls(text): #输入respone,寻找网址,返回列表urls
    url_pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    urls = url_pattern.findall(text)
    return urls


# 打开文件并读取内容
with open('tvboxdcjx.json', 'r', encoding='utf-8') as file:
    for line in file:
        lineurl=line.strip()
        extract_urls(lineurl) 
        for url in extract_urls(lineurl):
            print(url)
```
